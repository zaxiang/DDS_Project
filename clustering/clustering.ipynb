# Preparing Data

```python
df = pd.read_csv('data/new_health.csv')
df = df.drop(columns = ['Unnamed: 0'])
df = df.set_index('CASEID')

# dropping columns related to suicide 
df = df.drop(['suic_thought_pst_12month','suic_thought_pst_yr','suic_plan','suic_attempt','whodas_score'],axis=1)

df.head()
```

```python
df.shape
```

```python
# Removing suicide_risk from df 
suic_risk = df.suicide_risk
df = df.drop('suicide_risk',axis=1)
```

```python
# Replacing all floats & ints to categorical 
df = df.applymap(lambda x: 'Yes' if x == 1 else x)
df = df.applymap(lambda x: 'No' if x == 0 else x)
# Setting all values to type category 
df = df.astype('category')
```

# Clustering

```python
# Import module for linear algebra
import numpy as np
# Import module for data visualization
from plotnine import *
import plotnine
# Data visualization with matplotlib
import matplotlib.pyplot as plt
# Use the theme of ggplot
plt.style.use('ggplot')
# Import module for k-protoype cluster
from kmodes.kmodes import KModes
```

```python
# Convert dataframe to matrix
dfMatrix = df.to_numpy()
dfMatrix
```

# Elbow Method

```python
# Choosing optimal K
cost = []
for cluster in range(1, 10):
    try:
        kmodes = KModes(n_jobs = -1, n_clusters = cluster, init = 'Huang', random_state = 0)
        kmodes.fit_predict(dfMatrix)
        cost.append(kmodes.cost_)
        print('Cluster initiation: {}'.format(cluster))
    except:
        break
```

```python
# Converting the results into a dataframe and plotting them
df_cost = pd.DataFrame({'Cluster': range(1, 10), 'Cost': cost})
# Data viz
plotnine.options.figure_size = (8, 4.8)
fig = (
    ggplot(data = df_cost)+
    geom_line(aes(x = 'Cluster',
                  y = 'Cost'))+
    geom_point(aes(x = 'Cluster',
                   y = 'Cost'))+
    geom_label(aes(x = 'Cluster',
                   y = 'Cost',
                   label = 'Cluster'),
               size = 10,
               nudge_y = 1000) +
    labs(title = 'Optimal number of cluster with Elbow Method')+
    xlab('Number of Clusters k')+
    ylab('Cost')+
    theme_minimal()
)
```

```python
ggsave(plot=fig, filename='elbow_method.png')
```
