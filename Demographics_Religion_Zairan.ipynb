### Demographics Columns (everything is categorical)

- CATAG6 - AGE CATEGORY RECODE (6 LEVELS)
- HEALTH - OVERALL HEALTH / HEALTH2: OVERALL HEALTH RECODE
- IRSEX - GENDER
- IRMARIT - MARITAL STATUS
- EDUCCAT2 - EDUCATION LEVEL (MORE GENERAL)
- NEWRACE2 - RACE
- EMPSTATY - IMPUTATION REVISED EMPLOYMENT STATUS



```python
demographics_dict = {
'EMPSTATY_dic' : {1: 'E_full_time', 
                2: 'E_part_time', 
                3: 'E_Unemployed', 
                4: 'E_Other', 
                5: 'E_youth'},
'NEWRACE2_dic' : {1: 'White', 
                2: 'Black/Afr_Am', 
                3: 'Native_Am/AK_Native', 
                4: 'Native_HI/Other_Pac_Isl', 
                5: 'Asian', 
                6: 'more_than_one_race', 
                7: 'Hispanic'},
'EDUCCAT2_dic' : {1: 'Less_than_high_school', 
                2: 'High_school_graduate', 
                3: 'Some_college', 
                4: 'College_graduate', 
                5: 'Edu_youth'},
'IRMARIT_dic' : {1: 'Married', 
               2: 'Widowed', 
               3: 'Divorced_Separated', 
               4: "Not_Married"},
'IRSEX_dic' : {1: 'Male', 2: 'Female'},
'HEALTH_dic' : {1: 'H_Excellent', 2: 'H_Very_good', 3: 'H_Good', 4: 'H_Fair', 5: 'H_Poor'}, #94 dont know, 97 refused
'CATAG6_dic' : {1: 'A-12-17', 2: 'A-18-25', 3: 'A-26-34', 4: 'A-35-49', 5: 'A-50-64', 6: 'A-65'}
}
```

```python
demographics_2012 = drug_2012[['CASEID', 'CATAG6', 'HEALTH', 'IRSEX', 'IRMARIT', 'EDUCCAT2', 'NEWRACE2', 'EMPSTATY']]
```

```python
for col in demographics_2012.columns:
    if col != 'CASEID':
        demographics_2012[col] = demographics_2012[col].replace(demographics_dict[col+'_dic'])
```

### Missingness: 8 (94, 97) in HEALTH, 8347 (99) in IRMARIT, 8347 (99) in EMPSTATY

```python
#only run this once, it will overwrite the df
fill_values_health = np.random.choice(demographics_2012[(demographics_2012['HEALTH'] != 94) & (demographics_2012['HEALTH'] != 97)]['HEALTH'], 8)
fill_values_IRMARIT = np.random.choice(demographics_2012[demographics_2012['IRMARIT'] != 99]['IRMARIT'], 8347)
fill_values_EMPSTATY = np.random.choice(demographics_2012[demographics_2012['EMPSTATY'] != 99]['EMPSTATY'], 8347)

demographics_2012.loc[(demographics_2012['HEALTH'] == 94) | (demographics_2012['HEALTH'] == 97), 'HEALTH'] = fill_values_health
demographics_2012.loc[demographics_2012['IRMARIT'] == 99, 'IRMARIT'] = fill_values_IRMARIT
demographics_2012.loc[demographics_2012['EMPSTATY'] == 99, 'EMPSTATY'] = fill_values_EMPSTATY

```

```python
demographics_2012.columns = ['CASEID', 'Age', 'Health', 'Sex', 'Marriage', 'Education', 'Race', 'Employment']
```

```python
demographics_2012 = demographics_2012.set_index('CASEID')
```

```python
demographics_2012
```

```python
religion_2012.columns = ['CASEID', 'religion_services', 'religion_important', 'religion_influence_decision', 'religion_share_important']
```

```python
religion_demographics_2012 = demographics_2012.merge(religion_2012, left_on='CASEID', right_on='CASEID', how='inner')
```

```python
religion_demographics_2012.to_csv('data/religion_demographics_2012.csv')
```
