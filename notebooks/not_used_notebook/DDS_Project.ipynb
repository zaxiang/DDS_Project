# Creating Religious Features

```python
# Column names of Religious Features
relgion_df = drug[['SNRLGSVC','SNRLGIMP','SNRLDCSN','SNRLFRND']]

# Renaming Religious Features to be more readeable
religion_df = relgion_df.rename({'SNRLGSVC':'Past_12_months',
                                 'SNRLGIMP':'Important',
                                 'SNRLDCSN':'Influence',
                                 'SNRLFRND':'Friends_Share'},axis=1)
religion_df.shape
```

```python
# Values in dataset we want to drop from df 
drop_values = [94,97,98,99]

# dropping all rows in df that contain any values in drop_values
religion_df = religion_df[(religion_df.Past_12_months.isin(drop_values) == False) & 
            (religion_df.Important.isin(drop_values) == False) & 
            (religion_df.Influence.isin(drop_values) == False) & 
            (religion_df.Friends_Share.isin(drop_values) == False)]

religion_df.shape
```
